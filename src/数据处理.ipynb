```python
import numpy as np
import os
import math
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
def Collect_Runing_Data(processes, number, version=1,runtime=5):
    # 判断参数是否合法
    assert processes>=1
    assert number>=100

    time_collection=[]

    
    # 收集参数
    for i in range(runtime):
        result=os.popen('mpiexec -n %d ..\\bin\sieve%d.exe %d'%(processes,version,number)).read()
        # print(result)
        time=float(result[len(result)-10:])
        # print(time)
        time_collection.append(time)
    
    
    time_array=np.array(time_collection)
    mean=time_array.mean()

    print("版本:%d,进程数：%d,范围在%d以内，运行次数:%d,平均运行时间：%f"%(version,processes,number,runtime,mean))
    return mean
```

```python
def Broadcast_Time(processors, number,times):
    # 判断参数是否合法
    assert processors>=1
    assert number>=100

    time_collection=0
    counts=0
    for i in range(times):   
        result=os.popen(r'mpiexec -n %d ..\\bin\broadcast.exe %d'%(processors,number),"r").readlines()
        for line in result:
            if line[0:14]=='BroadcastTimes':
                index=line.find(",")
                try:
                    counts+=int(line[15:index])
                    time_collection+=float(line[-9:-2])
                except:
                    continue                
    mean=time_collection/counts
    lamb=mean/math.log(processors,10)
    lamb_ceil=mean/math.ceil(math.log(processors,10))
    print("进程数：%d,范围在%d以内，运行次数:%d,Broadcast时间：%f,lambda: %f ,lambda_ceil: %f"%(processors,number,times,mean,lamb,lamb_ceil))
    return mean
```

```python
def Runtime(n=100000000,times=20,processors=16,dataPath="../Data/RunTime_1e8_p16.npy"):
    # 收集数据
    RunTime=np.zeros([4,processors])
    for i in range(4):
        for j in range(processors):
            RunTime[i][j]= Collect_Runing_Data(j+1,n,i+1,times)
    
    # 保存数据
    try:
        np.save(dataPath,RunTime)
    except:
        np.save("temp.npy",RunTime)
```

```python
def Version_2_3():
    RunTime=np.zeros([5,21])

    for i in range(5):
        for j in range(21):
            n=1000000000/(10**(i+1))
            data1 = Collect_Runing_Data(j+1,n,2,20) 
            data2 = Collect_Runing_Data(j+1,n,3,20)
            RunTime[i][j] =(data1-data2)/data1

    np.save("../Data/Version_2_3.npy",RunTime)
```

```python
def RuntimePlot(dataPath,n,times,imagepath):
    # 载入数据
    Runtime=np.load(dataPath)
    font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc")  # 依赖于Windows平台
    processor=Runtime.shape[1]
    
    # 画图
    plt.figure(figsize=(20,8),dpi=80)
    x = np.array([i for i in range(1,processor+1)])
    for i in range(Runtime.shape[0]):
        plt.plot(x,Runtime[i,:],label='version-%d'%(i))
    
    # 添加描述信息
    plt.xlabel("处理器个数",fontproperties=font,fontsize="large")
    plt.ylabel("平均运行时间（单位:秒)（n=%s,times=%d)"%(n,times),fontproperties=font,fontsize="large")
    plt.title("素数筛选",fontproperties=font,fontsize="xx-large")

    plt.grid(True)
    plt.legend()
    plt.savefig(imagepath)
```

```python
def RuntimeImprovement(dataPath,n,times,imagepath):
    # 载入数据
    Runtime=np.load(dataPath)
    font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc")
    
    # 计算提升百分比
    version_1=Runtime[0,:]
    version_2=Runtime[1,:]
    version_3=Runtime[2,:]
    version_4=Runtime[3,:]
    Improvement_1=[(version_1[i]-version_2[i])/version_1[i] for i in range(len(version_1))]
    Improvement_2=[(version_1[i]-version_3[i])/version_1[i] for i in range(len(version_1))]
    Improvement_3=[(version_1[i]-version_4[i])/version_1[i] for i in range(len(version_1))]
    
    # 画图
    processor=Runtime.shape[1]
    x = np.array([i for i in range(1,processor+1)])
    plt.figure(figsize=(20,10),dpi=80)
    plt.plot(x,Improvement_1,label='version2-version1')
    plt.plot(x,Improvement_2,label='version3-version1')
    plt.plot(x,Improvement_3,label='version4-version1')


    # 添加描述信息
    plt.xlabel("处理器个数",fontproperties=font,fontsize="large")
    plt.ylabel("百分比",fontproperties=font,fontsize="large")
    plt.title("各版本性能提升百分比(n=%s,times=%d)(相比于version1)"%(n,times),fontproperties=font,fontsize="xx-large")
    plt.ylim((0,1))

    plt.grid(True)
    plt.legend()
    plt.savefig(imagepath)
    # plt.show()
```
