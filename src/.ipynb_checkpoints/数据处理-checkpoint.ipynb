```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
Runtime=np.load("RunTime.npy")
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc") 
```

```python
x = np.array([i for i in range(1,17)])
y1=Runtime[0,:]
y2=Runtime[1,:]
y3=Runtime[2,:]
y4=Runtime[3,:]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y1,label='version-1')
plt.plot(x,y2,label='version-2')
plt.plot(x,y3,label='version-3')
plt.plot(x,y4,label='version-4')

# 添加描述信息
plt.xlabel("处理器个数",fontproperties=font,fontsize="large")
plt.ylabel("平均运行时间（n=1e8,times=5)",fontproperties=font,fontsize="large")
plt.title("素数筛选",fontproperties=font,fontsize="xx-large")

plt.grid(True)
plt.legend()
plt.savefig("RunTime.png")
# plt.show()
```

```python
Version_2_3=np.load("Version_2_3.npy")
x = np.array([i for i in range(1,22)])
y1=Version_2_3[0,:]
y2=Version_2_3[1,:]
y3=Version_2_3[2,:]
y4=Version_2_3[3,:]
y5=Version_2_3[4,:]
plt.figure(figsize=(20,10),dpi=80)
plt.plot(x,y1,label='1e8')
plt.plot(x,y2,label='1e7')
plt.plot(x,y3,label='1e6')
plt.plot(x,y4,label='1e5')
plt.plot(x,y5,label='1e4')

# 添加描述信息
plt.xlabel("处理器个数",fontproperties=font,fontsize="large")
plt.ylabel("Time:(version_2 - version_3)/version_2",fontproperties=font,fontsize="large")
plt.title("版本3性能提升百分比",fontproperties=font,fontsize="xx-large")

plt.grid(True)
plt.legend()
plt.savefig("Version_2_3.png")
# plt.show()
```

## 各版本性能提升

```python
version_1=Runtime[0,:]
version_2=Runtime[1,:]
version_3=Runtime[2,:]
version_4=Runtime[3,:]
```

```python
Improvement_1=[(version_1[i]-version_2[i])/version_1[i] for i in range(len(version_1))]
Improvement_2=[(version_1[i]-version_3[i])/version_1[i] for i in range(len(version_1))]
Improvement_3=[(version_1[i]-version_4[i])/version_1[i] for i in range(len(version_1))]
```

```python
x = np.array([i for i in range(1,17)])
plt.figure(figsize=(20,10),dpi=80)
plt.plot(x,Improvement_1,label='version2-version1')
plt.plot(x,Improvement_2,label='version3-version1')
plt.plot(x,Improvement_3,label='version4-version1')


# 添加描述信息
plt.xlabel("处理器个数",fontproperties=font,fontsize="large")
plt.ylabel("百分比",fontproperties=font,fontsize="large")
plt.title("各版本性能提升百分比(n=1e8)(相比于version1)",fontproperties=font,fontsize="xx-large")
plt.ylim((0,1))

plt.grid(True)
plt.legend()
plt.savefig("Improvement.png")
# plt.show()
```
